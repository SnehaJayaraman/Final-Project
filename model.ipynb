```python
def draw_missing_data_table(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
draw_missing_data_table(data)
```

```python
from scipy import stats
import numpy as np
out = np.abs(stats.zscore(data._get_numeric_data()))
out
```

```python
data= data[(out < 3).all(axis=1)]
print(data.shape)
```

```python
RT={'Yes':1,'No':0}
data.RainToday = [RT[item] for item in data.RainToday]
data.RainTomorrow = [RT[item] for item in data.RainTomorrow]
```

```python

columns = ['WindGustDir', 'WindDir3pm', 'WindDir9am']
for col in columns:
    print(np.unique(data[col]))
data = pd.get_dummies(data, columns=columns)
data.head()
```

```python
data.columns
```

```python
y = data.RainTomorrow
X = data.drop('RainTomorrow', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
```

```python
print(X_train.head())
print(X_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
logreg = LogisticRegression(random_state=0)
logreg.fit(X_train,y_train)
pred = logreg.predict(X_test)
accuracy = accuracy_score(y_test,pred)
print('Accuracy :',accuracy)
print('Accuracy in %:',accuracy*100,'%')
```
